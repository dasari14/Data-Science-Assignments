```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Company names
companies = [
    "Allied Signal", "Bankers Trust", "General Mills", "ITT Industries",
    "J.P.Morgan & Co.", "Lehman Brothers", "Marriott", "MCI",
    "Merrill Lynch", "Microsoft", "Morgan Stanley", "Sun Microsystems",
    "Travelers", "US Airways", "Warner-Lambert"
]

# Corresponding Measure X values
measure_x = [
    24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39,
    40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00
]

# Plotting the data
plt.bar(companies, measure_x, color='skyblue')
plt.xticks(rotation=90,ha ='center')
plt.plot(figsize = (40,30))
plt.xlabel('Company')
plt.ylabel('Measure X')
plt.title('Measure X for Different Companies')
plt.show()
```

```python
df = pd.DataFrame(measure_x)
mean_value = df.mean()
std_dev= df.std()
variance= df.var()
print("Mean:", mean_value)
print("variance:", variance)
print("standard deviation :",std_dev)
```

```python
from scipy.stats import norm


mu = 45  # mean
sigma = 8  # standard deviation


time_allotted = 60


time_work_begins = 10

# Calculate the adjusted time for the normal distribution
adjusted_time = time_allotted - (time_work_begins + mu)

# Calculate the z-score
z_score = adjusted_time / sigma

# Calculate the probability using the cumulative distribution function (cdf)
probability = 1 - norm.cdf(z_score)

print("Probability that the service manager cannot meet his commitment:", probability)
```
